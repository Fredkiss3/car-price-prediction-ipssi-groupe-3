```python
import pandas as pd
import sqlite3
import os
```

```python
df = pd.read_csv(f'annoncesAll.csv')
```

```python
df
```

```python
all_brands = df.brand.str.upper().unique()
all_brands
```

```python
df.model.str.lower().unique()
```

```python
brand_models = df[['brand', 'model']]
brand_models
```

```python
for brand in all_brands:
    print(brand)
```

```python
all_colors = df.color.str.lower().unique()
all_colors
```

```python
db = sqlite3.connect(os.path.abspath('..') + '/db.sqlite3')
```

```python
for brand in all_brands:
    sql = f"INSERT INTO api_carbrand(name) VALUES ('{brand}');"
    print(f'inserting brand : {brand}')
    print(f'[SQL] : {sql}')
    db.execute(sql)
    db.commit()
```

```python
# with engine.connect() as conn:
cursor = db.execute("SELECT * FROM api_carbrand")
# print data using the cursor object
for result in cursor:
    _id, value = result
    print(f'{_id}: {value}')
```

```python
cursor = db.execute("SELECT * FROM api_carbrand where name = 'OPEL'")
# for result in  cursor:
_id, value = list(cursor)[0]
print(f'{_id}: {value}')
```

```python
for brand, model in brand_models.values:
    cursor = db.execute(f"SELECT * FROM api_carbrand where name = '{brand}'")
    brand_id, _ = list(cursor)[0]
    sql = f"INSERT INTO api_carmodel(name, brand_id) VALUES ('{model}', {brand_id});"
    print(f'inserting model : {model} for brand {brand}')
    print(f'[SQL] : {sql}')
    try:
        db.execute(sql)
        db.commit()
    except Exception as e:
        print(f"Error when inserting model : '{model}' for brand '{brand}'")
        print(e)
```

```python
cursor = db.execute("SELECT * FROM api_carmodel")
# print data using the cursor object
for result in cursor:
    _id, value = result[0], result[1]
    print(f'{_id}: {value}')
```

```python
for color in all_colors:
    sql = f"INSERT INTO api_carcolors(color) VALUES ('{color}');"
    print(f'inserting color : {color}')
    print(f'[SQL] : {sql}')
    db.execute(sql)
    db.commit()
```
